```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('C:/Chanwanich/NIA_e_commerce/Knowlage center/test_difference_group.csv')
df
```

```python
# Data visualization before the test using subplots
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

sns.histplot(df["Score_goup_1"], color="blue", ax=axes[0])
axes[0].set_title("Group 1")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

sns.histplot(df["Score_goup_2"], color="red", ax=axes[1])
axes[1].set_title("Group 2")
axes[1].set_xlabel("Value")

plt.suptitle("Data Distribution of Group 1 and Group 2")
plt.show()
```

```python
# Perform the t-test
t_stat, p_value = stats.ttest_ind(df["Score_goup_1"], df["Score_goup_2"])
# Set the significance level
alpha = 0.05
```

```python
# Print results
print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < alpha:
    print("There is a significant difference between the two groups.")
else:
    print("There is no significant difference between the two groups.")

# Visualize the results and hypothesis
plt.figure(figsize=(8, 6))
sns.barplot(x=["Score_goup_1", "Score_goup_2"], y=[np.mean(df["Score_goup_1"]), np.mean(df["Score_goup_2"])], yerr=[stats.sem(df["Score_goup_1"]), stats.sem(df["Score_goup_2"])], capsize=0.1)
plt.ylabel("Mean")
plt.title("Comparison of Group Means with 95% Confidence Intervals")
plt.show()
```
